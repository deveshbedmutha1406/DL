```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/mnistdataset/creditcard.csv")
df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
obj = StandardScaler()
X = df.drop(['Class', 'Time'], axis=1)
Y = df['Class']
```

```python
X['Amount'] = obj.fit_transform(X['Amount'].values.reshape(-1, 1))
X.describe()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, stratify=Y, test_size=0.3)
xtrain.shape
```

```python
ytrain = ytrain.astype(bool)
ytest = ytest.astype(bool)
```

```python
normal_train = xtrain[~ytrain]
normal_test = xtest[~ytest]
normal_train.shape
```

```python
ytrain.value_counts()
```

```python
fraud_train = xtrain[ytrain]
fraud_test = xtest[ytest]
```

```python
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model, Sequential
```

```python
class AutoEncoder(Model):
  def __init__(self, output_dim, ldim=8):
    super().__init__()
    self.encoder = Sequential([
        Dense(64, activation='relu'),
        Dropout(0.1),
        Dense(32, activation='relu'),
        Dropout(0.1),
        Dense(16, activation='relu'),
        Dropout(0.1),
        Dense(ldim, activation='relu'),
        Dropout(0.1),

    ])
    self.decoder = Sequential([
        Dense(16, activation='relu'),
        Dropout(0.1),
        Dense(32, activation='relu'),
        Dropout(0.1),
        Dense(64, activation='relu'),
        Dropout(0.1),
        Dense(output_dim, activation='relu'),
        Dropout(0.1),
    ])

  def call(self, inputs):
    en = self.encoder(inputs)
    de = self.decoder(en)
    return de
```

```python
model = AutoEncoder(output_dim=normal_train.shape[1])
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
h = model.fit(normal_train, normal_train, validation_data=(normal_test, normal_test), epochs=11)
```

```python
preds = model.predict(xtrain)
mse = np.mean(np.power(xtrain-preds, 2), axis=1)
err_df = pd.DataFrame({'recon_err': mse, 'class': ytrain})
err_df.groupby('class').describe()
```

```python
thres = 0.753869 + 3.197587
```

```python
test_preds = model.predict(xtest)
mse1 = np.mean(np.power(xtest-test_preds, 2), axis=1)
ypred = [(lambda er: 1 if er >= thres else 0)(er) for er in mse1]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(ytest, ypred))
```

```python
print(classification_report(ytest, ypred))
```
